```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 

import tensorflow as tf
from tensorflow.keras.models import Sequential 
from tensorflow.keras.layers import Dense, Flatten, Conv2D, Dropout, BatchNormalization, MaxPool2D
from tensorflow.keras.callbacks import ModelCheckpoint
from tensorflow.keras.datasets import cifar10
```

```python
(train_data, train_labels), (test_data, test_labels) = cifar10.load_data()
```

```python
train_data.shape
```

```python
label_to_names = {0:'airplane',1:'automobile',2:'bird',3:'cat',4:'deer',5:'dog',
                  6:'frog',7:'horse',8:'ship',9:'truck'}
```

```python
train_data = train_data/255.0 # Normalization
test_data = test_data/255.0
```

```python
def random_sample(X,y):
  total = X.shape[0]
  random = np.random.choice(total,10)
  random_data = X[random]
  random_label = y[random]

  for i in range(10):
    plt.subplot(2,5,i+1)
    plt.imshow(random_data[i])
    plt.xlabel(label_to_names[random_label[i][0]],color='g')
    plt.xticks(ticks=[])
    plt.yticks(ticks=[])
  plt.show()
random_sample(train_data,train_labels)
```

```python
def mlp():
  model = Sequential([
               Flatten(input_shape=(32,32,3)),
               Dense(256,activation='relu',kernel_initializer='he_normal'),
               Dense(256,activation='relu',kernel_regularizer='l2'),
               BatchNormalization(),
               Dense(128,activation='relu',kernel_regularizer='l2'),
               Dense(128,activation='relu'),
               Dropout(.5),
               Dense(32,activation='relu'),
               Dense(32,activation='relu'),
               Dropout(.5),
               Dense(16,activation='relu'),
               Dense(10,activation='softmax')
  ])
  return model 

mlp_model = mlp()
mlp_model.summary()
```

```python
mlp_model.compile('adam',loss='sparse_categorical_crossentropy',metrics=['acc'])
filepath = 'mlp/'
model_checkpoint = ModelCheckpoint(save_weights_only=True,save_freq='epoch',filepath=filepath+'checkpoint_{epoch:02d}')

def get_lr(epoch):
  if epoch < 30:
    return .001
  else:
    return .001 * (tf.math.exp(0.1*(30-epoch)))



history = mlp_model.fit(train_data,train_labels, batch_size=256,epochs=150,shuffle=True,
                        validation_split=.1,verbose=2,callbacks=[tf.keras.callbacks.LearningRateScheduler(get_lr),model_checkpoint])
```

```python
def plot(param):
  plt.plot(history.history[param])
  plt.plot(history.history[f'val_{param}'])
  plt.xlabel('Epoch')
  plt.ylabel(param)
  plt.title(f'{param} vs.Epoch')
  plt.legend(['Training','Validation'], loc='upper right')
  plt.show()
plot('loss')
```

```python
plot('acc')
```

```python
# evaulation
mlp_model.evaluate(test_data,test_labels,batch_size=256,verbose=2)
```

```python
def get_cnn_model():
  model = Sequential([
                  Conv2D(32,kernel_size=5,padding='same',activation='relu',kernel_initializer='he_normal',input_shape=(32,32,3)),
                  Conv2D(32,activation='relu',kernel_size=5,padding='same'),
                  BatchNormalization(),
                  Conv2D(128,activation='relu',kernel_size=5),
                  Conv2D(128,activation='relu',kernel_size=5),
                  MaxPool2D(2,2),
                  Dropout(.6),
                  Conv2D(128,activation='relu',kernel_size=5,padding='same'),
                  Conv2D(32,activation='relu',kernel_size=5,padding='same'),
                  Flatten(),
                  Dense(110, activation='relu'),
                  Dropout(.6),
                  Dense(16,activation='relu'),
                  Dense(10,activation='softmax')
  ])
  return model

model_cnn = get_cnn_model()
model_cnn.summary()
```

```python
model_cnn.compile('adam',loss='sparse_categorical_crossentropy',metrics=['acc'])

model_checkpoint = ModelCheckpoint(save_weights_only=True,save_freq='epoch',filepath=filepath+'../cnn/checkpoint_{epoch:02d}')

def get_clr(epoch):
  if epoch < 50:
    return .001
  else:
    return .001 * (tf.math.exp(0.1*(50-epoch)))

history = model_cnn.fit(train_data,train_labels,epochs=100,
          batch_size=250,verbose=2,
          validation_split=.1, callbacks=[model_checkpoint,tf.keras.callbacks.LearningRateScheduler(get_clr)])
```

```python
plot('loss')
```

```python
plot('acc')
```

```python
model_cnn.evaluate(test_data,test_labels,batch_size=250,verbose=2)
```

```python
model = mlp()
model.load_weights(filepath+'checkpoint_150')
```

```python
def show(model):
  total = test_data.shape[0]
  random = np.random.choice(total,5)
  random_data = test_data[random]
  random_labels = test_labels[random]
  predictions = model.predict(random_data)
  fig, axes = plt.subplots(5,2,figsize=(16,12))
  fig.subplots_adjust(hspace=.4,wspace=-0.2)
  
  for i, (prediction, image, label) in enumerate(zip(predictions,random_data,random_labels)):
    axes[i,0].imshow(np.squeeze(image))
    axes[i,0].get_xaxis().set_visible(False)
    axes[i,0].get_yaxis().set_visible(False)
    axes[i,0].text(10.,-1.5,f'{label_to_names[label[0]]}')

    axes[i,1].set_xticks(np.arange(len(prediction)))
    axes[i,1].bar(np.arange(len(prediction)),prediction)
    axes[i,1].set_title(f'Categorical distribution, Model prediction: {label_to_names[np.argmax(prediction)]}')
  plt.show()
```

```python
show(model)
```
